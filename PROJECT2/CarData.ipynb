Given a data point, the algorithm attempts to classify the data point with one of the classes it knows

```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
CarData=pd.read_csv("car.data")
CarData
```

Preprocessing is used to convert non numerical data into numerical data

```python
le=preprocessing.LabelEncoder()
buying=le.fit_transform(list(CarData["buying"]))
maint=le.fit_transform(list(CarData["maint"]))
door=le.fit_transform(list(CarData["door"]))
persons=le.fit_transform(list(CarData["persons"]))
lug_boot=le.fit_transform(list(CarData["lug_boot"]))
safety=le.fit_transform(list(CarData["safety"]))
cls=le.fit_transform(list(CarData["class"]))
```

x is going to be the features and y the label i.e what we are predicting

zip makes a tuple object of all the values of the variables

```python
predict="class"
```

```python
X=list(zip(buying,maint,door,persons,lug_boot,safety))
y=list(cls)
```

```python
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.1)
```

```python
x_train, y_train
```

```python
model=KNeighborsClassifier(n_neighbors=9)
model.fit(x_train,y_train)
acc=model.score(x_test,y_test)
acc
```

Now to see what the data points are and what the prediction is.

```python
predicted=model.predict(x_test)
names=['unacc','acc','good','vgood']

for x in range(len(predicted)):
    print("Predicted: ",names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
    #To show the distance between the data points
    n=model.kneighbors([x_test[x]],9,True)
    print("N;", n)
```

KNN measures distance between the data point to be classified and the rest of the data and gets the k number of neighbors that are closest to it. And by which class has the highest number of neighbors, it adds that data point to that class.
